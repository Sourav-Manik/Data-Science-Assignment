```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_books = pd.read_csv("/content/book.csv")
df_books
```

```python
#Observation 1 - 10% support and 25% confidence value
```

```python
frequent_itemsets_ap1 = apriori(df_books, min_support = 0.15,use_colnames= True)
frequent_itemsets_ap1.head()
```

```python
print(len(frequent_itemsets_ap1))
```

```python
frequent_itemsets_ap1.sort_values("support", ascending = False).head()
```

```python
rules_support1 = association_rules(frequent_itemsets_ap1, metric="support", min_threshold=0.1)
rules_support1.head()
```

```python
print(len(rules_support1))
```

```python
rules_confidence1 = association_rules(frequent_itemsets_ap1, metric="confidence", min_threshold=0.25)
rules_confidence1.head()
```

```python
print(len(rules_confidence1))
```

```python
# Vizualizing data
plt.scatter(rules_confidence1['support'],rules_confidence1['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()

#strip plot
sns.stripplot(y= 'conviction', data=rules_confidence1,jitter=True)
plt.ylabel('conviction')
plt.show()
```
