# Company Data

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets  
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
```

```python
# importing data
df_Co = pd.read_csv('/content/Company_Data.csv')
```

```python
df_Co.head()
```

```python
df_Co.info()
```

```python
df_Co.describe()
```

```python
df_Co=pd.get_dummies(df_Co,columns=['ShelveLoc'])
```

```python
df_Co.head()
```

```python
# Considering different sales value, Good Sales >= 9.32, Average Sales >5.39 and <9.32, Poor Sales <=5.39
```

```python
# Classifying sales into categorical data type
sales_result = []
for value in df_Co["Sales"]:
    if value >= 9.32:
        sales_result.append("Good Sales")
    elif value < 9.32 and value > 5.39:
        sales_result.append("Average Sales")
    else:
        sales_result.append("Poor Sales")
       
df_Co["sales_result"] = sales_result   
print(df_Co)
```

```python
df_Co.info()
```

```python
label_encoder = preprocessing.LabelEncoder()
df_Co['sales_result']= label_encoder.fit_transform(df_Co['sales_result']) 
df_Co['Urban']= label_encoder.fit_transform(df_Co['Urban']) 
df_Co['US']= label_encoder.fit_transform(df_Co['US']) 
df_Co.head()
```

```python
x=df_Co.iloc[:,1:13]
y=df_Co['sales_result']
```

```python
print(x.head())
print(y.head())
```

```python
df_Co.sales_result.value_counts()
```

```python
colnames = list(df_Co.columns)
colnames
```

```python
# Splitting data into training and testing data set
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=40)
```

```python
# Ramdon forest
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(bootstrap=True, class_weight = None,criterion='gini',n_estimators=15,random_state = None)
```

```python
model_rf.fit(x_train,y_train)
```

```python
y_predit = model_rf.predict(x_test)
np.mean(y_predit==y_test)
```

```python
# import some data to play with
df = pd.read_csv('/content/Fraud_check.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df=pd.get_dummies(df,columns=['Marital.Status'])
```

```python
df.head()
```

```python
# Classifying taxable income into categorical data type
Fraud_Check = []
for value in df["Taxable.Income"]:
    if value <= 30000:
        Fraud_Check.append("Risky")
    else:
        Fraud_Check.append("Good")
       
df["Fraud_Check"] = Fraud_Check   
print(df)
```

```python
df.info()
```

```python
label_encoder = preprocessing.LabelEncoder()
df['Fraud_Check']= label_encoder.fit_transform(df['Fraud_Check']) 
df['Undergrad']= label_encoder.fit_transform(df['Undergrad']) 
df['Urban']= label_encoder.fit_transform(df['Urban']) 
df.head()
```

```python
df=df.drop('Taxable.Income',axis=1)
df.head()
```

```python
x=df.iloc[:,0:7]
y=df['Fraud_Check']
```

```python
print(x.head())
print(y.head())
```

```python
df.Fraud_Check.value_counts()
```

```python
colnames = list(df.columns)
colnames
```

```python
# Splitting data into training and testing data set
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.25,random_state=20)
```

```python
# Ramdon forest
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rf2 = RandomForestClassifier(bootstrap=True, class_weight = None,criterion='gini',n_estimators=15,random_state = None)
```

```python
model_rf2.fit(x_train,y_train)
```

```python
y_predit = model_rf2.predict(x_test)
np.mean(y_predit==y_test)
```
