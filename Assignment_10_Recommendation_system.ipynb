```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('/content/book.csv',encoding = 'unicode_escape')
```

```python
df[0:5]
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
#rename the columns
df = df.rename({'User.ID': 'UserID','Book.Title': 'Title','Book.Rating': 'Rating' }, axis=1)
```

```python
df.head()
```

```python
df.shape
```

```python
#number of unique users in the dataset
len(df.UserID.unique())
```

```python
len(df.Title.unique())
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
user_book_df = df.pivot_table(index='UserID',
                        columns='Title',
                        values='Rating').reset_index(drop=True)
```

```python
user_book_df
```

```python
user_book_df.index = df.UserID.unique()
```

```python
user_book_df
```

```python
user_book_df.shape
```

```python
#Impute those NaNs with 0 values
user_book_df.fillna(0, inplace=True)
```

```python
user_book_df.head()
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df.head()
```

```python
#Set the index and column names to user ids 
user_sim_df.index = df.UserID.unique()
user_sim_df.columns = df.UserID.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
